## Discapacidad total en Mexico

#### Mostrar la cantidad de personas con discapacidad  por colonia  en el municipio de Naucalpan 
Este dataset fuente descargado de la pagina oficial del <strong>INEGI</strong>

##### fuentes:https://www.inegi.org.mx/app/scitel/Default?ev=9
##### notacion: {  *  } datos reservados por confidencialidad

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### Dataframe  de personas con discapacidad en EdoMex

```python
path=r"C:\Users\ASUS\Documents\MexicoSocial\discapacidad_Mex.csv"
edoM_df=pd.read_csv(path)
edoM_df
```

##### Dataframe con la informacion de personas discapacitadas  del municipio Naucalpan de Juarez

```python
nroDiscNau=edoM_df[(edoM_df.NOM_MUN=='Naucalpan de Juárez') & (edoM_df.MUN==57)]
nroDiscNau
```

#### Total de personas con discapacidad en Naucalpan de Juarez

```python
personasDiscTotal=nroDiscNau.iloc[0:1,3:7]
personasDiscTotal.set_index('NOM_MUN')


```

#### Dataframe con numeros reservados  en personas con discapacidad

```python

datosReservados=nroDiscNau[nroDiscNau['PCON_DISC']=='*']
datosReservados
```

#### Dataframe con numeros no reservados en personas con discapacidad

```python
datosNoReservados=nroDiscNau[(nroDiscNau['PCON_DISC']!='*') & (nroDiscNau['LOC']!=0) &  (nroDiscNau['LOC']!=1)]
datosNoReservados
```

#### Total de personas con discapacidad "contables" de numeros no reservados

```python
totalDiscContables=datosNoReservados['PCON_DISC'].astype('int64').sum()
print(f"Total de personas con discapacidad contables: {totalDiscContables}")
```

```python
fig,ax=plt.subplots()
ax.plot(datosNoReservados['LOC'],datosNoReservados['PCON_DISC'])
plt.show()
```
