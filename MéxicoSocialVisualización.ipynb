```python
dfDiscCaminar=pd.read_csv(pathOne)
display(dfDiscCaminar.head())

dfDiscVisual=pd.read_csv(pathTwo)
display(dfDiscVisual.head())

dfDiscHablar=pd.read_csv(pathThree)
display(dfDiscHablar.head())

dfDiscOir=pd.read_csv(pathFour)
display(dfDiscOir.head())

dfDiscVestir=pd.read_csv(pathFive)
display(dfDiscVestir.head())

dfDiscRecordar=pd.read_csv(pathSix)
display(dfDiscRecordar.head())

```

```python
def cleanDataNau(data):
    df=data[(data['NOM_MUN']=='Naucalpan de Juárez') & (data['MUN']==57) & (data.iloc[:,-1]!='*') & (data.LOC!=0) &(data.LOC!=1)]
    return df  
```

```python
dfCaminar=cleanDataNau(dfDiscCaminar)
display(dfCaminar.head())

dfVisual=cleanDataNau(dfDiscVisual)
display(dfVisual.head())

dfHablar=cleanDataNau(dfDiscHablar)
display(dfHablar.head())

dfOir=cleanDataNau(dfDiscOir)
display(dfOir.head())


dfVestir=cleanDataNau(dfDiscVestir)
display(dfVestir.head())

dfRecordar=cleanDataNau(dfDiscRecordar)
display(dfRecordar.head())
```

```python
#column modify 'PCON_DISC'  as type 'int64'
dfCaminar[['PCDISC_MOT']] = dfCaminar[['PCDISC_MOT']].apply(pd.to_numeric)
dfCaminar.sort_values('PCDISC_MOT',inplace=True,ascending=False)
```

```python
dfVisual[['PCDISC_VIS']] = dfVisual[['PCDISC_VIS']].apply(pd.to_numeric)
dfVisual.sort_values('PCDISC_VIS',inplace=True,ascending=False)
```

```python
dfHablar[['PCDISC_LENG']] = dfHablar[['PCDISC_LENG']].apply(pd.to_numeric)
dfHablar.sort_values('PCDISC_LENG',inplace=True,ascending=False)
```

```python
dfOir[['PCDISC_AUD']] = dfOir[['PCDISC_AUD']].apply(pd.to_numeric)
dfOir.sort_values('PCDISC_AUD',inplace=True,ascending=False)
```

```python

dfVestir[['PCDISC_MOT2']] = dfVestir[['PCDISC_MOT2']].apply(pd.to_numeric)

dfVestir.sort_values('PCDISC_MOT2',inplace=True,ascending=False)
```

```python

dfRecordar[['PCDISC_MEN']] = dfRecordar[['PCDISC_MEN']].apply(pd.to_numeric)

dfRecordar.sort_values('PCDISC_MEN',inplace=True,ascending=False)
```

```python

def graphPlot(df,xName,yName,color,titleX,titleY,mainTitle):
    
    fig,ax=plt.subplots(figsize=(8,15))
    plotBar=sns.barplot(data=df,y=yName,x=xName,ax=ax,color=color)
    ax.set_xlabel(titleX)
    ax.set_ylabel(titleY)
    ax.set_title(mainTitle)
    for p in ax.patches:
        label='{:.0f}'.format(p.get_width())
        y=p.get_y()+p.get_height()/2
        x=p.get_x()+p.get_width()
        ax.annotate(label, (x, y),rotation=0, size = 9)
    return plotBar    
```

```python
plotDiscVisual=graphPlot(dfVisual,'PCDISC_VIS','NOM_LOC','yellow','Personas con discapacidad visual','Localidad','Población con discapacidad para ver, aun usando lentes')
```

```python
plotDisMotriz=graphPlot(dfCaminar,'PCDISC_MOT','NOM_LOC','red',' Población con discapacidad para motriz','Localidad','Población con discapacidad para caminar, subir o bajar ')
```

```python
plotDiscHablar=graphPlot(dfHablar,'PCDISC_LENG','NOM_LOC','pink',' Población con discapacidad del lenguaje','Localidad','Población con discapacidad para hablar o comunicarse')
```

```python
plotDiscOir=graphPlot(dfOir,'PCDISC_AUD','NOM_LOC','green',' Población con discapacidad para oir','Localidad','Población con discapacidad para oír, aun usando aparato auditivo')
```

```python
plotDisVestir=graphPlot(dfVestir,'PCDISC_MOT2','NOM_LOC','orange',' Población con discapacidad motriz','Localidad','Población con discapacidad para vestirse, bañarse o comer')
```

```python
plotDisMental=graphPlot(dfRecordar,'PCDISC_MEN','NOM_LOC','brown',' Población con discapacidad para recordar','Localidad','Población con discapacidad para recordar o concentrarse')
```
