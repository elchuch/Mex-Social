## Discapacidad total en Mexico

#### Mostrar la cantidad de personas con discapacidad  por colonia  en el municipio de Naucalpan 
Este dataset fuente descargado de la pagina oficial del <strong>INEGI</strong>

##### fuentes:https://www.inegi.org.mx/app/scitel/Default?ev=9
##### notacion: {  *  } datos reservados por confidencialidad

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as ans
```

#### Dataframe  de personas con discapacidad en EdoMex

```python
path=r"C:\Users\ASUS\Documents\MexicoSocial\discapacidad_Mex.csv"
edoM_df=pd.read_csv(path)
edoM_df
```

```python
print(edoM_df.dtypes)
```

##### Dataframe con la informacion de personas discapacitadas  del municipio Naucalpan de Juarez

```python
nroDiscNau=edoM_df[(edoM_df.NOM_MUN=='Naucalpan de Juárez') & (edoM_df.MUN==57)]
nroDiscNau
```

#### Total de personas con discapacidad en Naucalpan de Juarez

```python
personasDiscTotal=nroDiscNau.iloc[0:1,3:7]
personasDiscTotal.set_index('NOM_MUN')


```

#### Dataframe con numeros reservados  en personas con discapacidad

```python

datosReservados=nroDiscNau[nroDiscNau['PCON_DISC']=='*']
datosReservados
```

#### Dataframe con numeros no reservados en personas con discapacidad 

```python
datosNoReservados=nroDiscNau[(nroDiscNau['PCON_DISC']!='*') & (nroDiscNau['LOC']!=0) &  (nroDiscNau['LOC']!=1)]
datosNoReservados
```

```python
print(datosNoReservados.dtypes)
```

### Total de personas con discapacidad "contables" de numeros no reservados

```python
#creating the bar plot
plt.style.use('Solarize_Light2')
fig,ax=plt.subplots(figsize=(8,15))
ax.barh(datosNoReservados.NOM_LOC,datosNoReservados['PCON_DISC'].astype('int64'),color='blue')
# Add padding between axes and labels
ax.tick_params(axis='y',pad=10 ,labelcolor='black', labelsize='medium', width=5)
ax.tick_params(axis='x',pad=10,labelcolor='blue', labelsize='medium', width=10)
ax.set_xlabel('Personas con discapacidad',fontsize=12)
ax.set_ylabel('Localidades',fontsize=12)
ax.set_title('Total de personas con discapacidad (Naucalpan de Juárez) "contables"', 
             loc='center',fontsize=12)

plt.show()
```

```python
totalDiscContables=datosNoReservados['PCON_DISC'].astype('int64').sum()
print(f"Total de personas con discapacidad contables: {totalDiscContables}")
```

```python
print(datosNoReservados.dtypes)
```

```python
#column modify 'PCON_DISC'  as type 'int64'
datosNoReservados[['PCON_DISC']] = datosNoReservados[['PCON_DISC']].apply(pd.to_numeric) 
print(datosNoReservados.dtypes)
```

### localidad(es) con el máximo numero de discapacitados

```python
maximas=datosNoReservados[datosNoReservados['PCON_DISC']==datosNoReservados['PCON_DISC'].max()]
maximas
```

### localidad(es) con el mínimo numero de discapacitados

```python
minimas=datosNoReservados[datosNoReservados['PCON_DISC']==datosNoReservados['PCON_DISC'].min()]
minimas
```
